```python
drawing = cv2.imread('C:/Users/alvar/OneDrive/Alvaro/Circle detection/PDF_Prueba.png', cv2.IMREAD_COLOR)
plt.imshow(drawing)
plt.show()
```

```python
drawing_gray = cv2.cvtColor(drawing, cv2.COLOR_BGR2GRAY)
plt.imshow(drawing_gray, cmap='gray')
plt.show()
```

```python
drawing_binary = cv2.threshold(drawing_gray, 180, 255, cv2.THRESH_BINARY)[1]
plt.imshow(drawing_binary, cmap='gray')
plt.show()
```

```python
drawing_binary_cropped = drawing_binary[0:300,:]
plt.imshow(drawing_binary_cropped, cmap='gray')
plt.show()
```

```python
drawing_binary_blurred = cv2.GaussianBlur(drawing_binary_cropped, (3, 3),0)
plt.imshow(drawing_binary_blurred, cmap='gray')
plt.show()
```

```python
edges = cv2.Canny(drawing_binary_blurred, 30, 15)
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
detected_circles = cv2.HoughCircles(drawing_binary_blurred, cv2.HOUGH_GRADIENT, dp=3, minDist=20, param1 = 10, param2 = 5, minRadius = 0, maxRadius = 4) 

# Draw circles that are detected. 
if detected_circles is not None: 
  
    # Convert the circle parameters a, b and r to integers. 
    detected_circles = np.uint16(np.around(detected_circles)) 
  
    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
  
        # Draw the circumference of the circle. 
        cv2.circle(drawing, (a, b), r, (0, 255, 0), 2) 
  
        # Draw a small circle (of radius 1) to show the center. 
        #cv2.circle(drawing, (a, b), 1, (0, 0, 255), 3) 

        #plt.imshow(drawing) 
        #plt.show()
        #time.sleep(5)

        
    
    plt.imshow(drawing) 
    plt.show 
  
```

```python
# show the image, provide window name first
cv2.imshow('image window', drawing)
# add wait key. window waits until user presses a key
cv2.waitKey(0)
# and finally destroy/close all open windows
cv2.destroyAllWindows()
```
