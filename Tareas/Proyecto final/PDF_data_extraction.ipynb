```python
import pandas as pd
import fitz
import re
```

```python
def PDFtextextract (pdfs):
    #pdfs---->Input: list of PDF locations
    #pdf_text_list---->Returned value: List of PDF texts
    #pdf_page_number---->Returned value: List of number of pages
    #pdf_has_text---->Returned value: List of 1 & 0 where 1 = document has text
    #pdf_word_count---->Returned value: List of number of words
    #pdf_characters_count---->Returned value: List of number of characters

    doc_num = len(pdfs) #Variable to keep track of progress
    progress_tracker = 0 #Variable to keep track of progress

    #variables to be returned:
    pdf_text_list = []
    pdf_page_number = []
    pdf_has_text = []
    pdf_word_count = []
    pdf_characters_count = []

    #Iteration through all files
    for file in pdfs:

        try:
            text = ''
            read_pdf = fitz.open(file) #Opens PDF
            pages = len(read_pdf) #Saves page quantity
            pdf_page_number.append(pages)

            #Iteration through the document pages:
            for pagenum in range(pages):
                text_temp = read_pdf[pagenum].get_text() #gets text from a page
                text = text + text_temp #joining all pages text into one variable
            pdf_text_list.append(text)

            character_num = len(text) #gets number of characters
            pdf_characters_count.append(character_num)
            word_num = len(re.findall(r'\w+', text)) #gets number of words
            pdf_word_count.append(word_num)

            #Determines if the document contains text
            if character_num > 100:
                pdf_has_text.append(1)

            else:
                pdf_has_text.append(0)

            read_pdf.close() #closes document

        except:
            pdf_text_list.append('')
            pdf_has_text.append(0)
            pdf_characters_count.append(0)
            pdf_word_count.append(0)
            pdf_page_number.append(0)

        progress_tracker += 1 #Variable to keep track of progress
        print('Text extraction progres: ' + str(progress_tracker) + '/' + str(doc_num), end= '\r') #Variable to keep track of progress
    
    return pdf_text_list, pdf_page_number, pdf_has_text, pdf_word_count, pdf_characters_count
```

```python
def PDFtableextract(pdfs):
    #pdfs---->Input: list of PDF locations
    #pdf_table_count---->Returned value: List of number of tables

    doc_num = len(pdfs) #Variable to keep track of progress
    progress_tracker = 0 #Variable to keep track of progress

    #variable to be returned:
    pdf_table_count = []

    #Iteration through all files:
    for pdf in pdfs:
        pdf_table_num = 0

        try:
            read_pdf = fitz.open(pdf) #Opens PDF
            pages = len(read_pdf) #Saves page quantity

            #Iteration through the document pages:
            for page in range(pages):
                tables = read_pdf[page].find_tables() #searches for tables in the page
                table_num = len(tables.tables) #gets number of tables in the page
                pdf_table_num = pdf_table_num + table_num #Adds the tables per page

            pdf_table_count.append(pdf_table_num)

        except:
            pdf_table_count.append(0)

        progress_tracker += 1 #Variable to keep track of progress
        print('Table extraction progres: ' + str(progress_tracker) + '/' + str(doc_num), end= '\r') #Variable to keep track of progress

    return pdf_table_count
```

```python
def PDFimageextract(pdfs):
    #pdfs---->Input: list of PDF locations
    #pdf_image_count---->Returned value: List of number of images

    doc_num = len(pdfs) #Variable to keep track of progress
    progress_tracker = 0 #Variable to keep track of progress

    #variable to be returned:
    pdf_image_count = []

    #Iteration through all files:
    for pdf in pdfs:
        pdf_image_num = 0

        try:
            read_pdf = fitz.open(pdf) #Opens PDF
            pages = len(read_pdf) #Saves page quantity

            #Iteration through the document pages:
            for page in range(pages):
                images = read_pdf[page].get_images() #Searches for images in the page
                image_num = len(images) #gets number of images in the page
                pdf_image_num = pdf_image_num + image_num #Adds the images per page

            pdf_image_count.append(pdf_image_num)

        except:
            pdf_image_count.append(0)

        progress_tracker += 1 #Variable to keep track of progress
        print('Table extraction progres: ' + str(progress_tracker) + '/' + str(doc_num), end= '\r') #Variable to keep track of progress

    return pdf_image_count
```

```python
#Reads database with download links
Data = pd.read_csv('C:/Users/alvar/Documents/Maestria en Ciencia de datos/Aprendizaje profundo/Proyecto final/Specs_to_extract.csv')
Data
```

```python
#Extraction of pdf data
pdf_save_loc = Data['download_location']
texts, page_num, has_text, word_count, char_count = PDFtextextract(pdf_save_loc)
```

```python
#Extraction of pdf tables number
table_count = PDFtableextract(pdf_save_loc)
```

```python
#Extraction of pdf tables images
image_count = PDFimageextract(pdf_save_loc)
```

```python
#Creating new dataframe with the relevant data:
specs = pd.DataFrame(texts, columns = ['pdf_texts'])
specs
```

```python
#Adding relevant columns to the dataframe
specs['page_number'] = page_num
specs['table_number'] = table_count
specs['image_number'] = image_count
specs['word_number'] = word_count
specs['character_number'] = char_count
spec_in_url = Data['spec_in_url'].tolist()
specs['spec_in_url'] = spec_in_url
doc_type = Data['doc_type'].tolist()
specs['doc_type'] = doc_type
specs['has_text'] = has_text

specs
```

```python
#Keeping only files that contain text:
specs = specs.loc[specs['has_text'] == 1]
specs
```
