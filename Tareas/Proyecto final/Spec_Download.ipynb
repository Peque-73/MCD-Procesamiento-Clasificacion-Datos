```python
import os
import re
import requests
import pandas as pd
```

```python
def pdfdownload_pathlist (link_list, save_path):
    #link_list---->Input: list of PDF URL
    #save_path---->Input: string containing the path where the PDFs are going to be saved
    #path_list---->Returned value: List containing all the PDFs locations
    path_list = []
    for link in link_list:

        try:
            response = requests.get(link, stream=True) #Connects to the url
            pdf_name = re.sub('[<>:"?|*]', 'X', os.path.basename(link)) #Gets the pdf name from the url and substitutes soecial characters with an 'x'

            if re.search(r'.pdf\Z|.PDF\Z', pdf_name) == None: #Verifies if the pdf name contains .pdf, if not we add it
                pdf_save_path = save_path+pdf_name+'.pdf' 

            else: 
                pdf_save_path = save_path+pdf_name

            if response.status_code == 200:
            
                with open(pdf_save_path, 'wb') as pdf_file: #Indicates where the file needs to be saved and its name
                    pdf_file.write(response.content) #We get the content of the url and write it to a file that is being saved in the desired location
                    path_list.append(pdf_save_path) #Saves in a list the location where the file was saved
                    print(pdf_name + ' successfully saved in: ')
                    print(pdf_save_path)
                    
            else:
                print(pdf_name + ' could not be saved')
                path_list.append('file not downloaded') #Indicates if a specific file was not downloaded.

        except:
            print(pdf_name + ' could not be saved')
            path_list.append('file not downloaded')

    return path_list
```

```python
#Reads database with download links
Data = pd.read_csv('C:/Users/alvar/Documents/Maestria en Ciencia de datos/Aprendizaje profundo/Proyecto final/Specs_to_download.csv')
Data
```

```python
#Saves links in a list
links = Data['url_specsheet_competition'].to_list()

#Adress where PDFs will be saved.
PDF_save_path = 'C:/Users/alvar/Documents/Maestria en Ciencia de datos/Aprendizaje profundo/Proyecto final/PDFs proyecto/'
```

```python
#Downloading files:
PDF_save_loc = pdfdownload_pathlist(links, PDF_save_path)
```

```python
#Saving the download files location in the dataframe
Data_downloaded = Data.copy()
Data_downloaded['download_location'] = PDF_save_loc
Data_downloaded
```

```python
#Removing files that couldn´t be downloaded
Data_downloaded_clean = Data_downloaded.loc[Data_downloaded['download_location'] != 'file not downloaded']
Data_downloaded_clean
```

```python
#Removing files that where duplicated:
Data_downloaded_clean = Data_downloaded_clean.drop_duplicates(subset = 'download_location')
Data_downloaded_clean
```

```python
#Saving the dataframe to a csv file:
Data_downloaded_clean.to_csv('C:/Users/alvar/Documents/Maestria en Ciencia de datos/Aprendizaje profundo/Proyecto final/Specs_to_extract.csv', index = False)
```
